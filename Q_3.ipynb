```python
import random
import os
import cv2
import torch
import matplotlib.pyplot as plt
import numpy as np
from ultralytics import YOLO

def load_labels(label_path):
    """Load ground truth bounding boxes from a YOLO label file."""
    boxes = []
    with open(label_path, 'r') as file:
        for line in file.readlines():
            data = line.strip().split()
            class_id = int(data[0])
            x_center, y_center, width, height = map(float, data[1:])
            boxes.append((class_id, x_center, y_center, width, height))
    return boxes

def yolo_to_bbox(yolo_box, img_width, img_height):
    """Convert YOLO format (x_center, y_center, width, height) to (x_min, y_min, x_max, y_max)."""
    class_id, x_center, y_center, width, height = yolo_box
    x_min = int((x_center - width / 2) * img_width)
    y_min = int((y_center - height / 2) * img_height)
    x_max = int((x_center + width / 2) * img_width)
    y_max = int((y_center + height / 2) * img_height)
    return class_id, x_min, y_min, x_max, y_max

def draw_boxes(image, boxes, color, label):
    """Draw bounding boxes on the image."""
    for class_id, x_min, y_min, x_max, y_max in boxes:
        cv2.rectangle(image, (x_min, y_min), (x_max, y_max), color, 2)
        cv2.putText(image, label, (x_min, y_min - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)
    return image

def main(test_images_path, test_labels_path, model_path, num_samples=4):
    """Visualize ground truth and predicted bounding boxes on test images."""
    model = YOLO(model_path)
    test_images = [img for img in os.listdir(test_images_path) if img.endswith(('.tif'))]
    
    selected_images = random.sample(test_images, num_samples)
    
    fig, axes = plt.subplots(1,num_samples, figsize=(30, 10 * num_samples), dpi=50)  # Bigger images, one per row

    
    for idx, img_name in enumerate(selected_images):
        img_path = os.path.join(test_images_path, img_name)
        label_path = os.path.join(test_labels_path, img_name.replace('.tif', '.txt').replace('.tif', '.txt'))
        
        image = cv2.imread(img_path)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        img_height, img_width = image.shape[:2]
        
        # Load ground truth
        gt_boxes = load_labels(label_path)
        gt_boxes = [yolo_to_bbox(box, img_width, img_height) for box in gt_boxes]
        
        # Predict using YOLO model
        results = model(img_path,verbose=False)[0]
        pred_boxes = []
        for result in results.boxes.xyxy:
            x_min, y_min, x_max, y_max = map(int, result[:4])
            pred_boxes.append((0, x_min, y_min, x_max, y_max))
        
        # Draw ground truth (Green) and predictions (Blue)
        image = draw_boxes(image, gt_boxes, (0, 255, 0), 'GT')
        image = draw_boxes(image, pred_boxes, (0, 0, 255), 'Pred')
    
        
        axes[idx].imshow(image)
        axes[idx].axis('off')
        axes[idx].set_title(img_name)
    
    plt.tight_layout()
    plt.show()

# Example usage
TEST_IMAGES_PATH = "data/test/images"
TEST_LABELS_PATH = "data/test/labels"
MODEL_PATH = "runs/detect/train15/weights/best.pt"
main(TEST_IMAGES_PATH, TEST_LABELS_PATH, MODEL_PATH)
```

a) Compute mAP50 with supervision and compare with your implementation. What do you observe ?

```python
import supervision as sv
from supervision.metrics import MeanAveragePrecision
from ultralytics import YOLO
import glob
import numpy as np



img_dir = "data/test/images"
label_dir = "data/test/labels"
model = YOLO("runs/detect/train15/weights/best.pt")


gt = []
pt = []

for imgs in os.listdir(img_dir):
    img_path = os.path.join(img_dir, imgs)
    label_path = os.path.join(label_dir, imgs.replace('.tif', '.txt'))
    
    results = model(img_path, verbose=False)[0]
    pt.append(sv.Detections.from_ultralytics(results))

    curr_labels = []
    with open(label_path, 'r') as file:
        lines = file.readlines()
        for line in lines:
            data = line.strip().split()
            class_id = int(data[0])
            cx, cy, w, h = map(float, data[1:])
            x1 = (cx - w / 2) * 416
            y1 = (cy - h / 2) * 416
            x2 = (cx + w / 2) * 416
            y2 = (cy + h / 2) * 416
            curr_labels.append([class_id, x1, y1, x2, y2])
    curr_labels = np.array(curr_labels)
    gt.append(sv.Detections(xyxy=curr_labels[:, 1:], class_id=curr_labels[:, 0].astype(int)))


# Compute mAP
map_metric = MeanAveragePrecision()
map_metric.update(pt, gt)
map_result = map_metric.compute()
print("mAP50 obtained on test set using Supervision Library: {:.2f}".format(map_result.map50))
```
